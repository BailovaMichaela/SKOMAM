```python
import numpy as np
import matplotlib.pyplot as plt

# Definice časového intervalu a rychlosti jako funkce v(t) = 1 / (2 + t)
t_fine = np.linspace(0, 1.2, 1000)  # jemná mřížka pro vykreslení funkce
v_fine = 1 / (0.5 + t_fine)  # funkce rychlosti

t_coarse = np.linspace(0, 1.2, 4)  # hrubší mřížka pro lichoběžníkovou metodu
v_coarse = 1 / (0.5 + t_coarse)  # hodnoty rychlosti na hrubší mřížce

# Lichoběžníkové pravidlo - výpočet aproximace
delta_t = t_coarse[1] - t_coarse[0]
area_trapz = np.sum((v_coarse[:-1] + v_coarse[1:]) / 2 * delta_t)

# Vizualizace
plt.figure(figsize=(5, 3))
# Jemná mřížka
plt.plot(t_fine, v_fine, label="Funkce rychlosti", color="blue")
# Hrubá mřížka s lichoběžníkovou aproximací
plt.plot(t_coarse, v_coarse, label="Lineární aproximace", color="orange", linestyle="--", marker="o")
for i in range(len(t_coarse) - 1):
    plt.fill_between(
        [t_coarse[i], t_coarse[i + 1]], [v_coarse[i], v_coarse[i + 1]], color="skyblue", alpha=0.4, label="Aproximace plochy pod křivkou" if i == 0 else ""
    )

plt.title("Ukázka lichoběžníkového pravidla pro numerickou integraci")
plt.xlabel("Čas (t)")
plt.ylabel("Rychlost (v)")
plt.legend()
plt.grid()
plt.show()
```
