# Cvičení 3: Je to padělek?
*Obyčejnou diferenciální rovnicí 1. řádu* rozumíme rovnici tvaru
$$y'(t)= f(t,y(t)),$$
kde $f :\, \mathbb{R}^2 \rightarrow \mathbb{R}$ je zadaná funkce a $y:\, \mathbb{R} \rightarrow \mathbb{R}$ je "hledaná" funkce. *Řešením této rovnice* *na otevřeném intervalu $(a,b)$*, kde $a<b$, rozumíme každou funkci $\bar{y}:\, (a,b) \rightarrow \mathbb{R}$, takovou, že pro všechna $t \in (a,b)$ platí
$$\bar{y}'(t)=f(t,\bar{y}(t)).$$

Úlohu, která se skládá z hledání řešení obyčejné diferenciální rovnice 1. řádu, která má navíc splňovat tzv. počáteční podmínku $y(t_0)=y_0$, $t_0 \in \mathbb{R},\, y_0 \in \mathbb{R}$, nazýváme *Cauchyovou úlohou* a zapisujeme ji obecně takto:
$$
\left\{ \begin{array}{l} y'(t)= f(t,y(t)),\\
      y(t_0)=y_0. \end{array} \right.
$$

### Eulerova metoda
Eulerova metoda je opravdu jednoduchý způsob numerického řešení Cauchyových úloh. Výstup Eulerovy metody nám aproximuje řešení Cauchyovy úlohy na intervalu $\langle t_0, t_N \rangle$. Metoda využívá aproximace derivace funkce $y$ v bodě $t$ pomocí tzv. dopředné diference $y$ v tomto bodě
$$y'(t) \approx \frac{y(t+h) - y(t)}{h},$$
kde $h$ je "malé" a kladné.
Pro jistotu zde připomínáme definici derivace funkce $y$ v bodě $t$: $$
y'(t) \stackrel{\mathrm{def.}}{=} \lim\limits_{h \rightarrow 0} \frac{y(t + h) - y(t)}{h}.$$

Po jednoduché úpravě dostaneme
$$y(t+h) \approx y(t) + hy'(t).$$
Použijeme-li vztah popisující Cauchyovu úlohu, pak získáme 
$$
y(t+h) \approx y(t) + hf(t,y(t)).
$$
Dále zvolme "dostatečně malé" $h>0$ a sestrojme posloupnost
$$t_0,\ t_1 \stackrel{\mathrm{def.}}{=} t_0 + h,\ t_2 \stackrel{\mathrm{def.}}{=} t_0 + 2h,\ \ldots, \ t_N \stackrel{\mathrm{def.}}{=} t_0 + Nh$$
Označme pomocí $y_n$ aproximaci hodnoty přesného řešení $y(t_n)$. 

Ze vztahu $y(t+h) \approx y(t) + hy'(t)$ a po dodání počáteční podmínky z Cauchyovy úlohy dostaneme rekurzivní vztah

$$
y_0=y(t_0),\\[4mm]
y_{n+1} = y_n + hf(t_n,y_n),\ n=0, 1, \ldots , N,
$$

který použijeme pro numerické řešení Cauchyovy úlohy. Dá se ukázat, že chyba aproximace řešení pomocí Eulerovy metody v bodě $t_1$ je přímo úměrná druhé mocnině velikosti kroku $h$. Chyba aproximace řešení v bodě $t_N$ je přímo úměrná velikosti $h$.

### Určení přibližného stáří obrazu pomocí metody radioaktivního datování
*Metoda radioaktivního datování* využívá toho, že některé tzv. radioaktivní prvky jsou nestabilní a část jejich atomů se samovolně rozpadá na atomy jiných prvků. Experimenty bylo zjištěno, že rychlost rozpadu atomů radioaktivních prvků je přímo úměrná počtu těchto atomů. Pokud funkci udávající počet atomů radioaktivního prvku v čase $t$ v gramu látky označíme jako $N(t)$, pak výše zmíněnou závislost můžeme popsat diferenciální rovnicí
$$
N'(t)= - \lambda N(t),
$$
kde $\lambda$ je konstanta, která popisuje rychlost rozpadu atomů daného radioaktivního prvku. Tato konstanta je dána pro každý radioaktivní prvek tímto vztahem
$$
\lambda = \frac{\ln 2}{\text{poločas rozpadu prvku}} .
$$

Čas $t$ v našem modelu budeme měřit v rocích a jednotka konstanty $\lambda$ je v ${rok}^{-1}$.

### Určení stáří "Vermeerova" obrazu
*   Jak v roce 1967 poznat, které Vermeerovy obrazy jsou pravé a které jsou padělky?
*   Zkoumejme olovnatou bělobu používanou na Vermeerových obrazech.
*   Barva obsahuje oxid olovnatý, který obsahuje malé množství olova-210 a ještě méně radia-226.
*   Ve chvíli, kdy je barva obsahující oxid olovnatý vyrobena, začnou se atomy olova-210 velmi rychle rozpadat s poločasem rozpadu 22 let a množství olova-210 v této barvě klesá.
*   Naopak vzniká malé množství olova-210 rozpadem radia-226 (a prvků, které následují v rozpadové řadě za ním). Poločas rozpadu radia-226 je 1600 let.

Tento proces můžeme popsat následující Cauchyovou úlohou
$$
\left\{ \begin{array}{l} N'(t)= -\lambda N(t) + r(t),\\[2mm]
      N(0)=N_0, \end{array} \right.
$$
kde $N(t)$ je funkce udávající počet atomů olova-210 v čase $t$ v gramu látky, $r(t)$ je funkce udávající počet atomů olova-210, které vzniknou v čase $t$ v gramu oxidu olovnatého za rok.

Protože poločas rozpadu radia-226 je 1600 let a metodu radioaktivního datování chceme použít pro rozpoznání stáří obrazů, které měly v roce 1967 buď přibližně 300 let nebo 20 let, můžeme funkci $r(t)$ považovat za konstantní. Pak $r(t)=r=\text{konst.}$ a předcházející úlohu můžeme nahradit úlohou
$$
\left\{ \begin{array}{l} N'(t)= -\lambda N(t) + r,\\[2mm]
      N(0)=N_0. \end{array} \right.
$$

Řešením předcházející úlohy jsme schopni, pokud známe počet atomů olova-210 v gramu oxidu olovnatého v době výroby olovnaté běloby, určit stáří obrazu, na kterém je tato barva použita. K řešení této úlohy můžeme použít Eulerovu metodu, se kterou jste se seznámili dříve. V naší úloze počet atomů olova-210 v gramu oxidu olovnatého v době výroby  barvy bohužel neznáme. I přesto jsme schopni rozlišit obraz jehož stáří je 300 let od obrazu, který má 20 let. Je totiž známo, jaké bývají koncentrace radioaktivního olova-210 v rudách, ze kterých se vyrábí oxid olovnatý. Na Zemi není známo ložisko, které by obsahovalo rudu s koncentrací větší než $5 \cdot 10^{11}$ atomů olova-210 v gramu rudy. Proto můžeme zjistit, pokud známe potřebné parametry, zda je možné, aby bylo stáří obrazu 300 let.

```python
import matplotlib.pyplot as plt # knihovna pro vykreslování grafů
import numpy as np # matematická knihovna
```

Nejprve si vytvoříme funkci, která pro zadané $N(t)$, kde $t=1967$, a $r$ určí, zda je obraz s těmito parametry starý 300 nebo 20 let.

```python
def stari_obrazu(N_t,r):
    # Funkce zjisti, zda je obraz se zadanymi parametry stary 20 ci 300 let.
    # Pokles poctu atomu Pb-210 v 1 gramu oxidu olovnateho popisujeme 
    # rovnici N'(t) = -lambda*N(t) + r s pocatecni podminkou N(0) = N0 na <0,t_N> 
    #
    # N_t  ... pocet atomu Pb-210 v 1 gramu oxidu olovnateho v case t
    # r ... pocet atomu Pb-210, ktere vzniknou v 1 gramu oxidu olovnateho za 1 rok rozpadem Ra-226 
    #
    # t ... zjistene stari obrazu (20 nebo 300 let)

    p_r = 22 # polocas rozpadu (v letech)
    l = np.log(2) / p_r

    n = 10**5 # pocet deleni <0,tN>

    critical = 5e11
    N_critical = critical*np.ones(n+1)

    # test pro stari 300 let
    tN = 300
        
    h = tN / n
    tt = np.linspace(0,tN,n+1)
    N = np.zeros(n+1)
    N[0] = N_t
       
    for i in range(n): 
      N[i+1] = N[i] - h*(-l*N[i] + r)
 
        
    if N[n] <= critical:
      t = 300
      print('Obraz je pravdepodobne stary 300 let a N(0) =', "%10.3E" % (N[n]),'.')
      
      N = np.flip(N) # prevraci poradi prvku pole
      
      plt.figure()
      plt.plot(tt,N)
      plt.plot(tt,N_critical,'g--')
      plt.yscale("log") # nastaveni logaritmicke skaly na ose y
      plt.xlabel('cas')
      plt.ylabel('koncentrace atomu')
      return t
        
    N = np.flip(N) # prevraci poradi prvku pole
        
    plt.figure()
    plt.plot(tt,N)
    plt.plot(tt,N_critical,'g--')
    plt.yscale("log") # nastaveni logaritmicke skaly na ose y
    plt.xlabel('cas')
    plt.ylabel('koncentrace atomu')
        
    # test pro stari 20 let
    tN = 20
   
    h = tN / n
    tt = np.linspace(0,tN,n+1)
    N = np.zeros(n+1)
    N[0] = N_t
    
    for i in range(n): 
      N[i+1] = N[i] - h*(-l*N[i] + r)
      
    if N[n] <= critical:
      t = 20
      print('Obraz je pravdepodobne stary 20 let a N(0) =', "%10.3E" % (N[n]),'.')

      N = np.flip(N) # prevraci poradi prvku pole

      plt.figure()
      plt.plot(tt,N)
      plt.plot(tt,N_critical,'g--')
      plt.yscale("log") # nastaveni logaritmicke skaly na ose y
      plt.xlabel('cas')
      plt.ylabel('koncentrace atomu')
      return t

    print('Nejsme schopni rozhodnout, zda je obraz stary 300 nebo 20 let.')
    return t
```
